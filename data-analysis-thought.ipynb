```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Whatsgoodly_Thought_Catalog_Influencers.csv')
df
```

```python
df_thought = df[df["Question"] == "What do you think when an influencer is obviously selling a product?"]
```

```python
# Filter the dataset to include only the "Global resuanlts" segment description
global_results = df_thought[df_thought['Segment Description'] == 'Global results']

# Set the social media platform as the index and extract the percentage values
platform_percents = global_results.set_index('Answer')['Percentage']

# Sort the values in descending order
platform_percents = platform_percents.sort_values(ascending=False)

# Create a horizontal bar chart
fig, ax = plt.subplots(figsize=(10,8))
platform_percents.plot(kind='bar', ax=ax)

# Add chart labels and title
ax.set_xlabel('Percentage of respondents')
ax.set_ylabel('Reaction')
ax.set_title('Reaction to Influence Selling a Product on Social Media Platform')
plt.show()
```
